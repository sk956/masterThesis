```python
# find the P1 identified person data  / find the p2 idnetified dataset / find the p3 identified dataset 

p1 = df.loc[df['identified_person'] == "P1"] 
p2 = df.loc[df['identified_person'] == "P2"] 
p3 = df.loc[df['identified_person'] == "P3"] 

# indexing to help with data analysis in R section.
p1.index =range(len(p1['identified_person']))
p2.index =range(len(p2['identified_person']))
p3.index =range(len(p3['identified_person']))
```

```python
# delete all the rows where the imgTimestamp is 0 

def delete_all_zero(pd): 
    list_drop = []
    for i in range(len(pd['img_timestamp'])): 
        if pd.iloc[i]['img_timestamp'] == 0:    
            list_drop.append(i)
    
    # delete the indexes 
    pd  = pd.drop(list_drop)
    return pd 
```

```python
p1 = delete_all_zero(p1)
p2 =  delete_all_zero(p2)
p3 = delete_all_zero(p3)

p1.index =range(len(p1['identified_person']))
p2.index =range(len(p2['identified_person']))
p3.index =range(len(p3['identified_person']))
```

```python
def emotions_col(pd) :
    pd['emotions'] =  (-1) *pd['confused'] + pd['calm'] + pd['happy'] - pd['disgusted'] + pd['angry'] - pd['fear'] + pd['sad'] + pd['surprised'] 
    
    return pd 
```

```python
# realized that the point graphs are cumulative 
def cum_emotions(pd): 
    pd['cum_emotions'] = 0.0
    for i in range(len(pd['emotions'])): 
        if i ==0 :
            pd.loc[i, 'cum_emotions'] = pd.iloc[i]['emotions']
        else: 
            pd.loc[i, 'cum_emotions'] = pd.iloc[i]['emotions'] +  pd.iloc[i-1]['cum_emotions']
        
    return pd
    
```

```python
""" Create a column that changes from ms to minutes  -> easier to look / view at the line graphs plot"""
def secs_to_min(pd): 
    pd['min_timestamp'] = pd['img_timestamp']/60 
    # change it to like minutes / seconds 
    # minutes and seconds 
    return pd
```

```python
""" seconds to minutes upgraded version to figureo out how to match the graph """
def secs_to_upMins(pd): 
    # minutes and seconds 
    pd['min_upgrade'] = 0.0
    for i in range(len(pd['img_timestamp'])): 
        
        mins = int(pd.iloc[i]['img_timestamp']/60) 
        secs = pd.iloc[i]['img_timestamp']%60
        newmins = mins + secs/100
        pd.loc[i,'min_upgrade'] = newmins
    return pd 
```

```python
""" main emotion and confidence level for each time line

calm	happy	disgusted	angry	fear	sad	surprised

""" 
def main_emotion(pd): 
    # main meotions 
    pd['main_emotion'] = 'emotion'
    pd['mainE_confidence'] = 0.0 
    for i in range(len(pd['img_timestamp'])): 
        # maximum of the values 
        max_e = max(pd.iloc[i]['calm'],pd.iloc[i]['happy'], pd.iloc[i]['disgusted'], pd.iloc[i]['angry'], 
                    pd.iloc[i]['fear'], pd.iloc[i]['sad'], pd.iloc[i]['surprised'])
        
        pd.loc[i, 'mainE_confidence'] = max_e
        
        if max_e == pd.iloc[i]['calm']: 
            pd.loc[i, 'main_emotion'] = 'calm' 
        
        if max_e == pd.iloc[i]['happy']: 
            pd.loc[i, 'main_emotion'] = 'happy' 
        if max_e == pd.iloc[i]['disgusted']: 
            pd.loc[i, 'main_emotion'] = 'disgusted' 
        if max_e == pd.iloc[i]['angry']: 
            pd.loc[i, 'main_emotion'] = 'angry' 
        if max_e == pd.iloc[i]['fear']: 
            pd.loc[i, 'main_emotion'] = 'fear' 
        if max_e == pd.iloc[i]['sad']: 
            pd.loc[i, 'main_emotion'] = 'sad' 
        if max_e == pd.iloc[i]['surprised']: 
            pd.loc[i, 'main_emotion'] = 'surprised'
        
    return pd 
        
```

```python
p1 = emotions_col(p1)
p1= cum_emotions(p1)
p1 = secs_to_min(p1)
p1 = secs_to_upMins(p1)
p1 = main_emotion(p1)

p2 = emotions_col(p2)
p2= cum_emotions(p2)
p2 = secs_to_min(p2)
p2 = secs_to_upMins(p2)
p2 = main_emotion(p2)

p3 = emotions_col(p3)
p3= cum_emotions(p3)
p3= secs_to_min(p3)
p3 = secs_to_upMins(p3)
p3 = main_emotion(p3)


# change the indexing 
```
