```python
def emotions_col(pd) :
    
#  confused- 
# happy + 
# disguated. - 
# angry + 
# sad + 
# fear + 
# calm 0 
# surprised  + 

    pd['emotions'] =  (-1) *pd['confused'] + pd['calm'] *(0) + pd['happy'] - pd['disgusted'] + pd['angry'] + pd['sad'] + pd['surprised'] + pd['fear']
    
    return pd 
```

```python
""" different forms of emotional analysis in the code:
changes made for the emotions: 
confused + 
happy + 
disguated. - 
angry - sad negative 
fear is not included here 
calm - 0 
surprised  + 
fear +
sad - 
"""
def prevEmotions_col(oneData): 
    oneData['emotion0'] = (+1) * oneData['fear']+ oneData['confused'] *(1) + oneData['happy'] + oneData['disgusted'] *(-1) + oneData['angry'] * (-1) + oneData['sad'] *(-1) + oneData['surprised']
    return oneData
```

```python
# realized that the point graphs are cumulative 
def cum_emotions(pd): 
    pd['cum_emotions'] = 0.0
    for i in range(len(pd['emotions'])): 
        if i ==0 :
            pd.loc[i, 'cum_emotions'] = pd.iloc[i]['emotions']
        else: 
            pd.loc[i, 'cum_emotions'] = pd.iloc[i]['emotions'] +  pd.iloc[i-1]['cum_emotions']
            
    pd['cum0_emotions'] =0.0 
    for i in range(len(pd['emotion0'])): 
        if i ==0 :
            pd.loc[i, 'cum0_emotions'] = pd.iloc[i]['emotion0']
        else: 
            pd.loc[i, 'cum0_emotions'] = pd.iloc[i]['emotion0'] +  pd.iloc[i-1]['cum0_emotions']
            
        
    return pd
    
```

```python
""" Create a column that changes from ms to minutes  -> easier to look / view at the line graphs plot"""
def secs_to_min(pd): 
    pd['min_timestamp'] = pd['img_timestamp']/60 
    # change it to like minutes / seconds 
    # minutes and seconds 
    return pd
```

```python
""" seconds to minutes upgraded version to figureo out how to match the graph """
def secs_to_upMins(pd): 
    # minutes and seconds 
    pd['min_upgrade'] = 0.0
    for i in range(len(pd['img_timestamp'])): 
        
        mins = int(pd.iloc[i]['img_timestamp']/60) 
        secs = pd.iloc[i]['img_timestamp']%60
        newmins = mins + secs/100
        pd.loc[i,'min_upgrade'] = newmins
    return pd 
```

```python
""" main emotion and confidence level for each time line

calm	happy	disgusted	angry	fear	sad	surprised

""" 
def main_emotion(pd): 
    # main meotions 
    pd['main_emotion'] = 'emotion'
    pd['mainE_confidence'] = 0.0 
    for i in range(len(pd['img_timestamp'])): 
        # maximum of the values 
        max_e = max(pd.iloc[i]['calm'],pd.iloc[i]['happy'], pd.iloc[i]['disgusted'], pd.iloc[i]['angry'], 
                    pd.iloc[i]['fear'], pd.iloc[i]['sad'], pd.iloc[i]['surprised'], pd.iloc[i]['confused'])
        
        pd.loc[i, 'mainE_confidence'] = max_e
        
        if max_e == pd.iloc[i]['calm']: 
            pd.loc[i, 'main_emotion'] = 'calm'
            
        if max_e == pd.iloc[i]['confused']: 
            pd.loc[i, 'main_emotion'] = 'confused'
        if max_e == pd.iloc[i]['happy']: 
            pd.loc[i, 'main_emotion'] = 'happy' 
        if max_e == pd.iloc[i]['disgusted']: 
            pd.loc[i, 'main_emotion'] = 'disgusted' 
        if max_e == pd.iloc[i]['angry']: 
            pd.loc[i, 'main_emotion'] = 'angry' 
        if max_e == pd.iloc[i]['fear']: 
            pd.loc[i, 'main_emotion'] = 'fear' 
        if max_e == pd.iloc[i]['sad']: 
            pd.loc[i, 'main_emotion'] = 'sad' 
        if max_e == pd.iloc[i]['surprised']: 
            pd.loc[i, 'main_emotion'] = 'surprised'
        
    return pd 
        
```

```python
""" What is this code doing? 

Calculating the average time stamp for each 


"""


from collections import defaultdict 

def createDict(p1): 
    
    P1Dict = defaultdict(float) 

    for i in range(len(p1['img_timestamp'])): 
        time  = p1.iloc[i]['img_timestamp']
        e =  p1.iloc[i]['cum_emotions']
        e0 = p1.iloc[i]['cum0_emotions'] 
        
        P1Dict[time] = [e0, e] 
    
    return P1Dict

def average_dict (p1,p2,p3): 
    """inputs- dictionary of each p1,p2,p3"""
    
    """edit code. - 3.22 for teams where there are only 2 participants rather than one"""
    
    if len(p3) == 0: 
        average0 = defaultdict(float)
        average  =defaultdict(float)
        for time in p1.keys(): 
            if time in p2: 
                # create the average default dict in this scenario 
                avg  = (p1[time][1] + p2[time][1])/2
                avg0 = (p1[time][0] + p2[time][0])/2

                average[time] = avg
                average0[time] = avg0 
    
    else: 
    
        average0 = defaultdict(float)
        average  =defaultdict(float)
        for time in p1.keys(): 
            if time in p2 and time in p3: 
                # create the average default dict in this scenario 
                avg  = (p1[time][1] + p2[time][1] + p3[time][1])/3
                avg0 = (p1[time][0] + p2[time][0] + p3[time][0])/3 

                average[time] = avg
                average0[time] = avg0 

    
    return average0, average

def add_avg(p1, avg0, avg): 
    """ add the average cum values for each time stamp""" 
    p1['average_e'] = 0.0 
    p1['average_e0'] = 0.0
    for i in range(len(p1['img_timestamp'])): 
        currT = p1.iloc[i]['img_timestamp'] 
        if currT in avg: 
            p1.loc[i,'average_e'] = avg[currT]  
            p1.loc[i,'average_e0'] = avg0[currT]  

    return p1
        

def average_emotions(p1, p2, p3): 
    
    P1Dict = createDict(p1) 
    P2Dict = createDict(p2) 
    P3Dict = createDict(p3) 
    
    avg0, avg = average_dict(P1Dict, P2Dict, P3Dict) 
    
    p1 = add_avg(p1, avg0, avg)
    p2 = add_avg(p2, avg0, avg) 
    p3 = add_avg(p3, avg0, avg)
        
    
    return p1,p2,p3
```
